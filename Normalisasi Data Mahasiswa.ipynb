```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
file_path = "C:/Users/User/DASI/Week 3/Data_Mahasiswa.csv"
df = pd.read_csv(file_path, delimiter=";")

# Menampilkan beberapa baris pertama dataset
df.head()
```

```python
# Memilih fitur numerik yang akan dinormalisasi
numeric_features = ["Programming Skill"]

# Min-Max Scaling (0-1)
min_max_scaler = MinMaxScaler()
df_minmax = df.copy()
df_minmax[numeric_features] = min_max_scaler.fit_transform(df[numeric_features])

# Standardization (Z-Score Normalization)
standard_scaler = StandardScaler()
df_standardized = df.copy()
df_standardized[numeric_features] = standard_scaler.fit_transform(df[numeric_features])

# Menampilkan hasil normalisasi pertama untuk dibandingkan
df_comparison = df[numeric_features].head(5).copy()
df_comparison["Programming Skill (MinMax)"] = df_minmax["Programming Skill"].head(5)
df_comparison["Programming Skill (Z-Score)"] = df_standardized["Programming Skill"].head(5)

df_comparison
```
