# Normalimenggunakan Min-Max Scaling dan Standardization (Z-Score Normalization) untuk beberapa fitur numerik

# Membaca dataset

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load dataset
file_path = "D:\MATA KULIAH\SEMESTER 6\DATA SCIENCE\Week 6\InggridCantik\SuperMarket Analysis.csv"
df = pd.read_csv(file_path)

# Menampilkan beberapa baris pertama dataset
df.head()
```

```python
# Memilih fitur numerik yang akan dinormalisasi
numeric_features = ["Unit price", "Quantity", "Tax 5%", "Sales", "cogs", "gross income", "Rating"]

# Min-Max Scaling (0-1)
min_max_scaler = MinMaxScaler()
df_minmax = df.copy()
df_minmax[numeric_features] = min_max_scaler.fit_transform(df[numeric_features])

# Standardization (Z-Score Normalization)
standard_scaler = StandardScaler()
df_standardized = df.copy()
df_standardized[numeric_features] = standard_scaler.fit_transform(df[numeric_features])

# Menampilkan hasil normalisasi pertama untuk dibandingkan
df_comparison = df[numeric_features].head(5).copy()
df_comparison["Unit price (MinMax)"] = df_minmax["Unit price"].head(5)
df_comparison["Unit price (Z-Score)"] = df_standardized["Unit price"].head(5)

df_comparison
```

Hasil Normalisasi
1. Min-Max Scaling
=>  Mengubah data ke skala 0 hingga 1.
    Contoh: Unit price awal 74.69 → menjadi 0.7188.

2. Z-Score Normalization (Standardization)
=>  Mengubah data sehingga memiliki mean = 0 dan standar deviasi = 1.
    Contoh: Unit price awal 74.69 → menjadi 0.7182.
